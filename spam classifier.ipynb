```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
import pickle
```

```python
df=pd.read_csv('mail_data.csv')
```

```python
print(df)
```

```python
data=df.where((pd.notnull(df)),'')
data.head(10)
```

```python
data.info()
```

```python
data.shape
```

```python
data.loc[data['Category']=='spam','Category',]=0
data.loc[data['Category']=='ham','Category',]=1
```

```python
x=data['Message']
y=data['Category']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)

x_train_features=feature_extraction.fit_transform(x_train)
x_test_features=feature_extraction.transform(x_test)

y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
print(x_train)
```

```python
print(x_train_features)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train_features,y_train)
```

```python
prediction_on_training_data=model.predict(x_train_features)
accuracy_on_training_data=accuracy_score(y_train,prediction_on_training_data)
cm=confusion_matrix(y_train,prediction_on_training_data)
```

```python
print('Accuracy on training data',accuracy_on_training_data)
print(cm)
print(classification_report(y_train,prediction_on_training_data))
```

```python
prediction_on_testing_data=model.predict(x_test_features)
accuracy_on_testing_data=accuracy_score(y_test,prediction_on_testing_data)
cm_test=confusion_matrix(y_test,prediction_on_testing_data)
```

```python
print("Accuracy on Testing Data",accuracy_on_testing_data)
print(cm_test)
print(classification_report(y_test,prediction_on_testing_data))
```

```python
with open('model_pickle','wb') as file:
    pickle.dump(model,file)
```

```python
with open('model_pickle','rb') as file:
    mp = pickle.load(file)
```
